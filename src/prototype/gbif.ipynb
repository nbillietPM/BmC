# Lookup taxon names

Lookup a single scientific names information

```python
testName = "Arabis soyeri"

pygbif.species.name_backbone(name=testName)
```

Look up a list of names 

```python
testNames = "Arabis soyeri, Bryum pseudotriquetrum, Carex appropinquata, Catoscopium nigritum, Cochlearia pyrenaica, Cratoneuron commutatum, Cratoneuron filicinum, Drepanocladus vernicosus, Epilobium davuricum, Eucladium verticillatum, Juncus triglumis, Philonotis calcarea, Saxifraga aizoides, Scorpidium revolvens"
namesList = testNames.split(",")
print(namesList)
```

```python
taxonInfo = list(map(lambda x: pygbif.species.name_backbone(name=x), namesList))
```
