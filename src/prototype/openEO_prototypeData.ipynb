The Copernicus Vegetation Indices collection has a summary of the following vegetation indices

* NDVI (Normalized Difference Vegetation Index), the difference between red band and near IR band. This is the most commonly used vegetation index
* PPI (Plant Phenology Index), monitor plant phenology—the timing of biological events in vegetation such as leaf-out, flowering, and senescence. It is specifically developed to provide biophysically meaningful information about vegetation dynamics, especially in ecosystems with pronounced seasonal patterns
* LAI (Leaf Area Index),  dimensionless quantity that characterizes plant canopies. It is defined as the one-sided green leaf area per unit ground surface area in broadleaf canopies
* FAPAR (Fraction of Absorbed Photosynthetically Active Radiation), fraction of the incoming solar radiation in the photosynthetically active radiation spectral region that is absorbed by a photosynthetic organism, typically describing the light absorption across an integrated plant canopy.
