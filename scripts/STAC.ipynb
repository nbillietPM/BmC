```python
from pystac_client import Client

# Use the public URL if you are working locally
stac_url = "https://catalogue.dataspace.copernicus.eu/stac" 
# OR use your internal URL if you are running this inside their cloud environment
# stac_url = "https://catalogue-svc.prod-catalogue:8250/stac"

client = Client.open(stac_url)

print("Datasets (Collections) available:")
for collection in client.get_collections():
    print(f"- {collection.id}: {collection.title}")
```
