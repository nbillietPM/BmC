```python
import openeo.processes as p

# 2. Load ONLY the LAI band
# We rely on the product's internal '255' nodata flag instead of external SCL
lai_cube = connection.load_collection(
    "TERRASCOPE_S2_LAI_V2",
    spatial_extent=spatial_extent,
    temporal_extent=temporal_extent,
    bands=["LAI_10M"]
)

# 3. Mask No-Data Values (255)
# We explicitly define the mask: True where data is 255 (Bad), False where data is valid.
# This replaces 255 with JSON 'null' (which is safe for math).
nodata_mask = (lai_cube == 255)
lai_clean = lai_cube.mask(nodata_mask, replacement=None)

# 4. Apply Scale Factor (0.04)
# Turn raw integers (e.g., 125) into physical LAI (e.g., 5.0)
# We do this AFTER masking so we don't accidentally scale the 255s.
lai_physical = lai_clean * 0.04

# 5. (Optional) Interpolation
# Since you had success with this in your manual script, we can add it back here.
# Now that 255s are proper 'nulls', this will work correctly.

lai_filled = lai_physical.apply_dimension(
    process="array_interpolate_linear", 
    dimension="t"
)

lai_layers = []
for label in ["min", "max", "median", "mean", "sd"]:
    month_stat = lai_filled.aggregate_temporal_period(period="month", reducer=label)
    labeled_cube = month_stat.rename_labels(dimension="bands", target=[label] )
    lai_layers.append(labeled_cube)

lai_final_cube = lai_layers[0]
for layer in lai_layers[1:]:
    lai_final_cube = lai_final_cube.merge_cubes(layer)


# 9. Execute
job = lai_final_cube.execute_batch(
    out_format="NetCDF", 
    title="LAI_Physical_Stats"
)
print(f"Job started: {job.job_id}")
```

## Make 1km, 5km and 10km maps for Europe

```python
europe_extent = {"west":-10.371094,"south":34.016242,"east":46.582031,"north":72.23551}
temporal_extent = ["2019-04-01", "2019-06-30"]
```

### 10KM grid

```python
lai_cube = connection.load_collection(
    "TERRASCOPE_S2_LAI_V2",
    spatial_extent=europe_extent,
    temporal_extent=temporal_extent,
    bands=["LAI_10M"]
)

# 3. Mask No-Data Values (255)
# We explicitly define the mask: True where data is 255 (Bad), False where data is valid.
# This replaces 255 with JSON 'null' (which is safe for math).
nodata_mask = (lai_cube == 255)
lai_clean = lai_cube.mask(nodata_mask, replacement=None)

# 4. Apply Scale Factor (0.04)
# Turn raw integers (e.g., 125) into physical LAI (e.g., 5.0)
# We do this AFTER masking so we don't accidentally scale the 255s.
lai_physical = lai_clean * 0.04

# 3. RESAMPLE SPATIALLY (The Key Step)
# We aggregate 10m -> 10km (10000 meters)
# method="average" ensures we take the mean of valid pixels
lai_10km = lai_physical.resample_spatial(
    resolution=10000, 
    projection=3857, # or your target CRS
    method="average" 
)

# 4. Interpolate Temporally (On the 10km Cube)
# Now we only interpolate the few remaining gaps
lai_filled = lai_10km.apply_dimension(
    process="array_interpolate_linear", 
    dimension="t"
)

# 5. Aggregate Monthly
month_stat = lai_filled.aggregate_temporal_period(period="month", reducer="mean")

# 6. Filter & Execute
# Filter strictly for May 2019 to remove the buffer months
final_result = month_stat.filter_temporal("2019-05-01", "2019-05-31")

job = final_result.execute_batch(
    out_format="NetCDF", 
    title="LAI_May_2019_10km"
)
print(f"Job started: {job.job_id}")
```
