```python
import xarray as xr
import sparse
import numpy as np
def load_sparse(ds_sparse: xr.Dataset) -> xr.Dataset:
    """
    Rebuild the sparse.COO from a pseudo‐dense dataset and return
    a one‐variable xr.Dataset using the original var_name.
    """
    # pull back the components
    coords = ds_sparse["coords"].values
    data   = ds_sparse["data"].values
    shape  = tuple(ds_sparse["shape"].values)
    dims   = list(ds_sparse["dims"].values)

    # rebuild sparse.COO
    s = sparse.COO(coords, data, shape=shape)

    # recover any real coords (e.g. time, x, y)
    coord_vars = {
        c: ds_sparse[c]
        for c in ds_sparse.coords
        if c not in ("coords", "data", "shape", "dims")
    }

    # read the original var name from attrs
    var_name = ds_sparse.attrs.get("__orig_var_name__", None)
    if var_name is None:
        raise KeyError("Dataset is missing the __orig_var_name__ attribute")

    # build the DataArray
    da = xr.DataArray(
        s,
        dims=dims,
        coords=coord_vars,
        name=var_name,
        attrs={k: v for k, v in ds_sparse.attrs.items() if k != "__orig_var_name__"}
    )

    # wrap back into a Dataset
    return da.to_dataset()
```

```python
import xarray as xr
import sparse
import numpy as np

def sparse_to_pseudodense(dataset: xr.Dataset, var_name: str) -> xr.Dataset:
    """
    (Re-use your existing function, but ensure it stores __orig_var_name__ in attrs.)
    """
    sparse_data = dataset[var_name].data  # sparse.COO
    ds_sparse = xr.Dataset({
        "coords": (("ndim", "nnz"), sparse_data.coords),
        "data":   ("nnz",            sparse_data.data),
        "shape":  ("ndim",           np.array(sparse_data.shape, dtype=np.int64)),
        "dims":   ("ndim",           list(dataset[var_name].dims)),
    })
    for c in dataset[var_name].coords:
        ds_sparse[c] = dataset[var_name].coords[c]
    ds_sparse.attrs["__orig_var_name__"] = var_name
    ds_sparse.attrs.update(dataset[var_name].attrs)
    return ds_sparse

def clone_tree_with_pseudodense(source: DataTree) -> DataTree:
    """
    Recursively clone `source`. Whenever a leaf node holds an xr.Dataset
    with a sparse.COO DataArray, replace that leaf’s .data with the
    pseudo-dense dataset returned by `sparse_to_pseudodense`.
    """
    # Leaf case
    if source.is_leaf:
        ds = source.data
        # only act if it’s an xr.Dataset containing a sparse.COO
        if isinstance(ds, xr.Dataset) and any(isinstance(ds[v].data, sparse.COO)
                                             for v in ds.data_vars):
            pseudo = sparse_to_pseudodense(ds)
            return DataTree(name=source.name, data=pseudo, children={})
        # otherwise, just copy the leaf unchanged
        return DataTree(name=source.name, data=ds, children={})
    
    # Internal node: recurse into each child
    new_children = {
        name: clone_tree_with_pseudodense(child)
        for name, child in source.children.items()
    }
    # carry over any non-leaf data (e.g. metadata on intermediate groups)
    data = getattr(source, "data", None)
    return DataTree(name=source.name, data=data, children=new_children)

```

```python
def sparse_to_pseudodense(dataset, var_name):
    """
    Convert one sparse DataArray into a "pseudo‐dense" xr.Dataset, but
    hang onto the original var_name so load_sparse never needs to be told.
    """
    sparse_data = dataset[var_name].data  # sparse.COO
    
    ds_sparse = xr.Dataset({
        "coords": (("ndim", "nnz"), sparse_data.coords),
        "data":   ("nnz",            sparse_data.data),
        "shape":  ("ndim",           np.array(sparse_data.shape, dtype=np.int64)),
        "dims":   ("ndim",           list(dataset[var_name].dims)),
    })

    # copy over the real coordinate variables
    for c in dataset[var_name].coords:
        ds_sparse[c] = dataset[var_name].coords[c]

    # save the original var name as a dataset attribute
    ds_sparse.attrs["__orig_var_name__"] = var_name

    # if you want, also carry over any top‐level attrs
    ds_sparse.attrs.update(dataset[var_name].attrs)
    return ds_sparse
```
