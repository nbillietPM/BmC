```python
import openeo
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
```

```python
# Establish a connection to openeo
connection = openeo.connect("openeofed.dataspace.copernicus.eu")
```

```python
# List collections available on the openEO back-end
connection.list_collection_ids()

# Get detailed metadata of a certain collection
connection.describe_collection("SENTINEL2_L2A")
```

```python
connection.authenticate_oidc()
```

Latitude (North): 50.9224° N

Latitude (South): 50.9224° N (for the southern part of the garden)

Longitude (East): 4.3005° E

Longitude (West): 4.2905° E

For the Sentinel 2 satellite the following bands code for 

* B02 (Blue) 
* B03 (Green)
* B04 (Red)

```python
datacubeMeise = connection.load_collection(
    "SENTINEL2_L2A",
    spatial_extent={"west": 4.2905, "south":  50.9224, "east": 4.3005, "north": 50.9224},
    temporal_extent = ["2021-02-01", "2021-04-30"],
    bands=["B02", "B03", "B04"],
    max_cloud_cover=85,
)
```
