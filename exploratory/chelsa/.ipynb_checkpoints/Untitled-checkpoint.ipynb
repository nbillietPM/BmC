The following models are available when utilizing CMIP6 ISIMIP3

    1) "GFDL-ESM4"
    2) "UKESM1-0-LL"
    3) "MPI-ESM1-2-HR"
    4) "IPSL-CM6A-LR"
    5) "MRI-ESM2-0"
    
These models each represent a diffierent simulation of the Earth's climùate based on different underlying assumptions and numerical methods.

In the context of CHELSA and climate modeling in general, ensemble members refer to different simulations from the same climate model that vary slightly in their initial conditions or parameters. These are used to assess internal variability of the climate system — i.e., the natural randomness in weather and climate that exists even under the same external forcing (like greenhouse gas concentrations).

CHELSA uses single ensemble members that are structurally represented as $r_ai_bp_cf_d$

    * r_a, realization index 'a'
    * i_b, initilization method 'b'
    * p_c, physics version 'c'
    * f_d, forcing index 'd'

the following members are available

    * r1i1p1f1 
    * r1i1p1f2 

The scenarios are greenhouse gas concentration trajectories that represent the different potential futures based on human activity

    1) ssp126, SSP1-RCP2.6 climate as simulated by the GCMs.
        * Sustainable development (low emissions)
    2) ssp370, SSP3-RCP7 climate as simulated by the GCMs.
        * Regional rivalry (high emissions)
    3) ssp585, SSP5-RCP8.5 climate as simulated by the GCMs.
        * Fossil-fueled development (very high emissions)

Each SSP-x.x combines:

    * SSP: A narrative about socioeconomic development.
    * x.x: Radiative forcing level (in W/m²) by 2100.

```python
scenarios = ['ssp126','ssp245','ssp370','ssp585']

model = 'MPI-ESM1-2-LR'
member = 'r1i1p1f1'
model_members = [(model, member)]

xmin = 2.362061
xmax = 6.602783
ymin = 49.535191
ymax = 51.415653
region = [xmin, xmax, ymin, ymax]

fefps = '2041-01-15'
fefpe = '2070-12-15'
fefp = [(fefps, fefpe)]
```

Download the bioclimatic variables based on the scenarios defined earlier and the model members, future period and region as set above. This whole process might take a while.


This function will run chelsa_cmip6.
The input parameters are given and the original ones as defined in the original package in ():

    * model_members a list of tuples defining the model (source_id) and the member (member_id)
    * scenarios a list of SSP scenarios (experiment_id)
    * region is a list of coordinates (xmin, xmax, ymin, ymax) defining the region to create the output for
    * fefp is a list of tuples defining the future periods (fefps and fefpe)
    * refp is a tuple defining the reference period (refps, refpe) - it defaults to '1981-01-15' - '2010-12-15'
    * output_folder is the output folder that will be created automatically and will host the generated output (output) - it defaults to 'output'

```python
chelsa_cmip6.run_chelsa_cmip6(model_members, scenarios, region, fefp, refp=('1981-01-15','2010-12-15'), output_folder='output')
```

Ensure your reference and future periods contain full years, e.g., from January 1 to December 31, not mid-month.
These mid-month dates might result in partial months being sliced off, depending on how CHELSA or CMIP6 are processed internally.

```python
chelsa_cmip6(activity_id='ScenarioMIP',
             table_id='Amon',
             experiment_id="ssp585", 
             institution_id='MPI-M', 
             source_id='MPI-ESM1-2-LR', 
             member_id='r1i1p1f1', 
             refps='1981-01-01',
             refpe='2010-12-31',
             fefps='2020-01-01',
             fefpe='2020-12-31',	
             xmin=5.3,
             xmax=10.4,
             ymin=46.0,
             ymax=47.5,
             output=os.getcwd()) 
```
