```python
import openeo
import openeo.processes as proc

# 1. SETUP & CONFIGURATION
# Connect to Copernicus Data Space Ecosystem (or your specific backend)
connection = openeo.connect("https://openeo.dataspace.copernicus.eu").authenticate_oidc()

target_resolution = 1000
target_crs = 3857
temporal_window = ["2022-12-01", "2023-12-31"]
spatial_bxl = {"west": 4.147339, "south": 50.665567, "east": 4.757767, "north": 50.970535, "crs":4326}
spatial_vl = {"west": 2.543335, "south": 50.677316, "east": 5.866699, "north": 51.508742, "crs":4326}
4.113281,50.673835,4.783447,50.968806
# Define Time Windows
# Bio needs a wider window for gap-filling; LC is static for the year
bio_window = ["2022-12-01", "2023-03-31"] 
lc_window  = ["2021-01-01", "2021-12-31"]

# ==============================================================================
# PART 1: THE MASTER CUBE (BIODIVERSITY)
# This cube defines the target 1km grid for everything else.
# ==============================================================================

# 1. Load Sentinel-2
s2_cube = connection.load_collection(
    "SENTINEL2_L2A",
    spatial_extent=spatial_bxl,
    temporal_extent=bio_window,
    bands=["B02", "B03", "B04", "B05", "B06", "B07", "B08", "B11", "B12", "SCL"]
)

# 2. Mask Clouds (Keep Veg(4), Bare(5), Water(6))
scl = s2_cube.band("SCL")
clean_mask = (scl == 4) | (scl == 5) | (scl == 6)
s2_masked = s2_cube.mask(~clean_mask)

# 3. Scale to Reflectance
bands_to_load = ["B02", "B03", "B04", "B05", "B06", "B07", "B08", "B11", "B12"]
scaled = {b: s2_masked.band(b) * 0.0001 for b in bands_to_load}
# Define natural log transformation for individual cubes
# We multiply by -1 as per the NDLI requirements
log_b11_neg = scaled["B11"].apply(lambda x: proc.ln(x) * -1)
log_b08_neg = scaled["B08"].apply(lambda x: proc.ln(x) * -1)


# 4. Calculate Indices (Just 2 examples for brevity, you can add all 12)
eps = 1e-6
min_ref = 0.01

indices = {
    # --- Standard Vegetation & Soil Indices ---
    # NDVI: Proxy for biomass and greenness
    "NDVI":  (scaled["B08"] - scaled["B04"]) / (scaled["B08"] + scaled["B04"] + eps),
    
    # EVI: Enhanced index for dense canopy (uses Blue band to reduce noise)
    "EVI":   2.5 * (scaled["B08"] - scaled["B04"]) / (scaled["B08"] + 6.0 * scaled["B04"] - 7.5 * scaled["B02"] + 1.0 + eps),
    
    # MSAVI: Soil-adjusted index for sparse/arid vegetation
    "MSAVI": (2 * scaled["B08"] + 1 - ((2 * scaled["B08"] + 1)**2 - 8 * (scaled["B08"] - scaled["B04"]))**0.5) / 2,

    # --- Red-Edge & Pigment Indices ---
    # NDRE: Chlorophyll content using Red-Edge 1
    "NDRE":  (scaled["B08"] - scaled["B05"]) / (scaled["B08"] + scaled["B05"] + eps),
    
    # REIP: Red-Edge Inflexion Point wavelength (Standard S2 linear method)
    "REIP":  705 + 35 * (((scaled["B04"] + scaled["B07"]) / 2) - scaled["B05"]) / (scaled["B06"] - scaled["B05"] + eps),
    
    # ARI: Anthocyanin Reflectance Index (stress indicator)
    "ARI": (
        # 1. Calculate the raw formula as before
        (1 / (scaled["B03"] + eps)) - (1 / (scaled["B05"] + eps))
    ).mask(
        # 2. Apply the "Singularity Mask"
        # If Green (B03) OR RedEdge (B05) is too dark, we mask the result (set to NaN).
        (scaled["B03"] < min_ref) | (scaled["B05"] < min_ref)
    ),
    
    # CRI: Carotenoid Reflectance Index
    "CRI": (
        (1 / (scaled["B02"] + eps)) - (1 / (scaled["B03"] + eps))
    ).mask(
        # SINGULARITY MASK
        # If either input band is too dark, the math is invalid.
        (scaled["B02"] < min_ref) | (scaled["B03"] < min_ref)
    ),

    # --- Water, Moisture & Disturbance ---
    # NDWI: Water bodies and surface wetness
    "NDWI":  (scaled["B03"] - scaled["B08"]) / (scaled["B03"] + scaled["B08"] + eps),
    
    # NDMI: Leaf moisture and canopy water stress
    "NDMI":  (scaled["B08"] - scaled["B11"]) / (scaled["B08"] + scaled["B11"] + eps),
    
    # NBR: Normalized Burn Ratio for fire severity and recovery
    "NBR":   (scaled["B08"] - scaled["B12"]) / (scaled["B08"] + scaled["B12"] + eps),

    # --- Functional Diversity Indices ---
    # FPAR: Fraction of Photosynthetically Active Radiation
    "FPAR":  (((scaled["B08"] - scaled["B04"]) / (scaled["B08"] + scaled["B04"] + eps)) * 1.25) - 0.1,
    
    # NDLI: Normalized Difference Lignin Index (uses natural logs)
    "NDLI": (log_b11_neg - log_b08_neg) / (log_b11_neg + log_b08_neg + eps)
}

# 5. Process: Gap Fill -> Resample to Master Grid -> Monthly Aggregation
bio_layers = []
for label, cube in indices.items():
    # Gap Fill
    filled = cube.apply_dimension(process="array_interpolate_linear", dimension="t")
    
    # MASTER GEOMETRY DEFINITION
    # This specific step defines the grid (X, Y) that the Land Cover must match later.
    resampled = filled.resample_spatial(resolution=1000, projection=3857, method="bilinear")
    
    # Aggregate to Month (Max value to avoid clouds)
    monthly_stat = resampled.aggregate_temporal_period(period="month", reducer="max")
    
    # Labeling
    labeled_cube = monthly_stat.add_dimension(name="bands", label=label, type="bands")
    bio_layers.append(labeled_cube)

# 6. Merge to create the final Biodiversity Object
bio_master_cube = bio_layers[0]
for layer in bio_layers[1:]:
    bio_master_cube = bio_master_cube.merge_cubes(layer)

# ==============================================================================
# PART 2: THE FOLLOWER CUBE (LAND COVER)
# This cube is forced to align with bio_master_cube.
# ==============================================================================

# 1. Load WorldCover
lc_cube = connection.load_collection(
    "ESA_WORLDCOVER_10M_2021_V2",
    spatial_extent=spatial_bxl, # Must allow covering the same area
    temporal_extent=lc_window
)

lc_classes = {
    "trees": 10,
    "shrubland": 20,
    "grassland": 30,
    "cropland": 40,
    "builtup": 50,
    "barren": 60,
    "snowIce": 70,
    "water": 80,
    "herbWetland":90,
    "mangroves":95,
    "mossLichen":100
}
lc_layers = []

for name, value in lc_classes.items():
    # 2. Binary Mask & Float Conversion
    # * 1.0 is CRITICAL to prevent integer rounding to zero
    binary_class = (lc_cube.band("MAP") == value)
    float_mask = binary_class * 1.0 
    
    # 3. SPATIAL ALIGNMENT
    # We use 'resample_cube_spatial' to force this data into the bio_master_cube grid.
    # It copies the extent, resolution, and pixel alignment exactly.
    lc_aligned = float_mask.resample_cube_spatial(
        target=bio_master_cube, 
        method="average" # Returns fraction 0.0-1.0
    )
    
    # 4. Labeling
    labeled_lc = lc_aligned.add_dimension(
            name="bands", 
            label=name, 
            type="bands"
        )
    lc_layers.append(labeled_lc)

# 5. Merge Land Cover layers
lc_final_cube = lc_layers[0]
for layer in lc_layers[1:]:
    lc_final_cube = lc_final_cube.merge_cubes(layer)

# ==============================================================================
# PART 3: EXECUTION
# ==============================================================================

print("Submitting Biodiversity Job...")
job_bio = bio_master_cube.execute_batch(out_format="NetCDF", title="Bio_Master_1km")

print("Submitting LandCover Job...")
job_lc = lc_final_cube.execute_batch(out_format="NetCDF", title="LC_Aligned_1km")
```
